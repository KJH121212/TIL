## 이후의 변수들 결측치 보간
하기전에 x,y좌표는 중요하기 때문에 카카오와 네이버지도의 크롤링을 통해 데이터 확보 - 결측치 보간  

- 모든 연속형 변수를 선형 보간을 때려버림
['전용면적', '계약년월', '계약일', '층', '건축년도', 'k-전체동수', 'k-전체세대수', 'k-연면적', 'k-주거전용면적', 'k-관리비부과면적', '건축면적', '주차대수']  
계약일은 추후에 제거할 예정. 기준금리 때문에 있어야 함.

전용면적, 계약년월, 층, 건축년도는 결측치가 없기 때문에 선형보간이 필요하지 않음.  
나머지는 결측치가 약 75%에 근접함. 모조리 버리고 싶으나... 쓸모가 있을지도 모르니 일단 남겨두기로함.  
남은 ['k-전체동수', 'k-전체세대수', 'k-연면적', 'k-주거전용면적', 'k-관리비부과면적', '건축면적', '주차대수']을 모두 선형보간하기에는 서로 상관관계가 너무 크다.  


위와 같이 'k-전체세대수', 'k-연면적', 'k-주거전용면적', 'k-관리비부과면적' 4개의 column은 상관 계수가 무척 큰 것을 확인 할 수 있다.  
따라서 KNN 기반 보간을 통해 보간을 진행 한 이후, 다시 상관관계를 확인한 후 연면적을 제외한 모든 column을 제거하는 방법을 생각해 보았다.  
문제점은 보간하는데 걸리는 시간이 더럽게 오래 걸린다는 것.  
n_neighbors를 좀 작게 하면 시간이 적게 들어갈지 모르겠음. 몇번 시도해 보면서 적당한 수준을 찾아봐야 할 듯함.
