해당 column 들이 결측치가 가장 적기 때문에 이를 이용해 파생변수를 만드는 것이 합리적이면서 좋은 판단이다.
확인 결과 traing_dataset 또한 동일한 column들이 결측치가 적게 존재하는 것을 확인했다.

k-전용면적별세대현황(60㎡이하)         869608  
k-전용면적별세대현황(60㎡~85㎡이하)     869608

두 column의 결측치가 동일 한 것을 보아 하나의 행에 동시에 값이 들어갈 수 있는 정보라는 것인데...  
해당 단지에서 전용면적이 60㎡ 이상 ~ 85㎡ 이하인 세대 수를 나타낼 확률이 높다.
사실 결측치가 많기 때문에 다 제외 해버리는게 옳아 보이긴 하지만... 고민해볼 문제

## column 정리

1. 위치 관련 정보
시군구: 아파트가 위치한 시/군/구 정보.  (동마다, 구마다 target 값을 기준으로 상중하 로 등급을 나눌수 있음. 파생병수 생성 후 제거?)  [보류]*  
번지: 해당 주소의 번지수.               (결국 제공하는 정보는 위치 정보이기 때문에 필요 없을 것 같음)                               [제거]*  
본번: 주소의 본번 정보.                 (결국 제공하는 정보는 위치 정보이기 때문에 필요 없을 것 같음)                               [제거]*  
부번: 주소의 부번 정보.                 (결국 제공하는 정보는 위치 정보이기 때문에 필요 없을 것 같음)                               [제거]*  
도로명: 해당 아파트의 도로명 주소.      (xy 좌표 크롤링에 사용하고 제거)                                                            [제거]*  
좌표X: 아파트의 X 좌표.                 (위치정보 1)                                                                                [유지]  
좌표Y: 아파트의 Y 좌표.                 (위치정보 2)                                                                                [유지]  

2. 아파트 정보 관련 컬럼
아파트명: 아파트 단지 이름.             (아파트 브랜드 명 얻기 위해 사용 후, 너무 다양한 값을 가지기 때문에 제거)                   [제거]*  
층: 거래된 아파트의 층수.               (필요한 정보인지 모르겠음)                                                                  [보류]  
건축년도: 아파트가 건축된 연도.         (신축인지 아닌지가 중요한 정보이기 때문에 필수)                                             [유지]*  
k-단지분류(아파트,주상복합등등): 단지의 유형(아파트, 주상복합 등). (결측치가 너무 많음)                                             [보류]
k-건설사(시공사): 아파트 건설을 담당한 시공사.                  (건설사 등급 매기기 가능)                                           [보류]
k-시행사: 아파트 개발을 주도한 시행사.                          (시행사 등급 매기기 가능)                                           [보류]
단지소개기존clob: 아파트 단지 소개 정보.                        (긴 줄글은 해당 모델에 알맞지 않음)                                 [제거]
k-전체동수: 단지의 전체 동(건물) 수.                            (필요한지 의문+상관계수 with 전체 세대수)                           [제거]
k-전체세대수: 단지의 총 세대 수.                                (세대수가 크다는 것은 아파트가 크다는 것을 의미)                    [유지]

3. 면적 및 세대 정보
전용면적(㎡): 해당 거래된 아파트의 전용면적.                    (무조건 남아 있어야 됨)                                             [유지]*
k-주거전용면적: 단지 내 주거 전용 면적.                         (상관계수 너무 크게 나옴)                                           [보류]
k-연면적: 아파트 단지의 총 연면적.                              (상관게수 너무 크게 나옴)                                           [제거]
k-관리비부과면적: 관리비 산정에 사용되는 면적.                  (상관계수 너무 크게 나옴)                                           [제거]
건축면적: 아파트 면적                                           (주거전용면적이랑 비슷할거 같은데 의외로 상관계수는 낮음)           [보류]


k-전용면적별세대현황(60㎡이하): 전용면적 60㎡ 이하 세대 수.                 해당 4개가 모두 합해져야 아파트의 전체 세대 수가 나옴   [보류]
k-전용면적별세대현황(60㎡~85㎡이하): 전용면적 60㎡~85㎡ 이하 세대 수.
k-85㎡~135㎡이하: 전용면적 85㎡~135㎡ 이하 세대 수.
k-135㎡초과: 전용면적 135㎡ 초과 세대 수.

4. 거래 및 계약 정보
계약년월: 거래된 계약의 연월.                       (기준금리 만들때도 사용할 수 있고, 물가의 차이 때문에 무조건 필요)              [유지]*
계약일: 계약이 체결된 날짜.                         (게약 일까지 필요한가 싶긴하지만...)                                            [보류]*
거래유형: 해당 거래의 유형.                         (중개거래/직거래 판단은 필요 없음)                                              [제거]*
등기신청일자: 등기 신청 날짜.                       (base로 제거됨)                                                                 [제거]*
해제사유발생일: 계약 해제 사유가 발생한 날짜.       (결측치가 많아서 제거됨)                                                        [제거]

5. 관리 및 운영 정보
k-관리방식: 단지 관리 방식.                         (자치관리/위탁관리/사업주체 관리 3개로 나뉨)(결측치가 많아 고민)                [제거]
k-복도유형: 복도 유형(개방형, 폐쇄형 등).           (애매)                                                                          [보류]
k-난방방식: 난방 방식(중앙, 개별 등).               (애매)                                                                          [보류]
경비비관리형태: 경비 관리 방식.(직영/위탁)          (관리비에 관한 거니까 필요하지 않을까...?)                                      [유지]
세대전기계약방법: 세대 전기 계약 방식.              (관리비에 관한 거니까 필요하지 않을까...?)                                      [유지]
청소비관리형태: 청소 관리 방식.                     (관리비에 관한 거니까 필요하지 않을까...?)                                      [유지]
관리비 업로드: 관리비 데이터 업로드 여부. (Y/N)     (관리비에 관한 거니까 필요하지 않을까...?)                                      [유지/제거]          

6. 단지 관련 날짜
k-사용검사일-사용승인일: 사용 검사일 또는 사용 승인일.          (datetime type인데 그냥 남기는게 나을듯?)                           [유지]
단지승인일: 해당 단지가 승인된 날짜.                            (이게 위에거 보다 최근)                                             [유지]
단지신청일: 단지 등록 신청 날짜.                                (승인일과 많이 차이 나 봤자 일주일 정도 되는듯?)                    [제거]
k-등록일자: 데이터가 등록된 날짜.                               (결측치가 너무 많아 제거)                                           [제거]
k-수정일자: 데이터가 마지막으로 수정된 날짜.                    (이거는 실거래가와는 상관이 없지요?)                                [제거]

7. 기타 정보
k-세대타입(분양형태): 분양/기타 로 나뉘어짐                     (애매...)                                                           [보류]
k-전화번호: 단지 관리 사무소 전화번호.                          (무조건 제거)                                                       [제거]
k-팩스번호: 단지 관리 사무소 팩스번호.                          (무조건 제거)                                                       [제거]
k-홈페이지: 단지 홈페이지 URL.                                  (무조건 제거)                                                       [제거]
고용보험관리번호: 고용보험 관리 번호.                           (무조건 제거)                                                       [제거]
주차대수: 단지 내 주차 가능 대수.                               (있으면 좋을듯. 결측치 문제)                                        [유지]
사용허가여부: 정확한 내용 유추 불가                                                                                                 [제거]
기타/의무/임대/임의=1/2/3/4: 기타 단지 분류 코드.               (이게 뭐지? 모르는데 결측치까지 많음)                               [보류]
