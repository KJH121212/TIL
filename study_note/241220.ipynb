# ML 추천 시스템
## 추천시스템 (Recommendation or Recommender System)
데이터를 활용해, 기하급수적으로 많은 옵션들 중에서 사람들이 원하는 것이 무엇인지 예측하고, 범위를 좁혀주고, 찾아주는 머신러닝의 한 분류
- 사용자 : User
- 아이템/콘텐츠 : 추천되는 대상

구성요소  
1. 활동기록 : 아이템을 구매, 특정 페이지 조회 여부, 페이지 체류시간 등과 같은 사용자의 모든 행동에 대한 기록
2. 피드백 : 활동 기록중 일부로, 특정 아이템에 대한 사용자의 선호, 또는 비선호를 드러낼 수 있는 행동들 (ex. YouTube의 좋아요 / Netflix의 관심작 등)
3. 유저 프로필 : 여러 활동기록과 피드백을 통해 수집한 유저에 대한 정보를 모아놓은 데이터

기업의 수익과 직접적으로 연관된 기술이므로 공부해야할 이유가 충분하다.


## 추천시스템 활용
1. 스트리밍 서비스 : 수많은 콘텐츠 중, 좋아할만한 콘텐츠 찾기.
2. 소션 네트워크 서비스 : 콘텐츠 피드(무한 스크롤링: Youtube shorts, 등), 유해정보 필터링
3. 교육 서비스 : 어떤 유형의 문제를 틀리는지 확인해 추천 분야 제공
4. 헬스케어 서비스: 환자에게 맞는 치료법 추천
5. 광고 프랫폼의 추천시스템: 관심 분야 물품

유저 프로필 : 여러 활동기록과 피드백을 통해 수집한 유저에 대한 정보를 모아놓은 데이터
아이템 프로필 : 다양한 방식으로 수집된 각 콘텐츠에 대한 정보들 - 영화의 장르, 감독이나 출연배우 등에 기반한 추천
유저 피드백 정보 : 다양한 방식으로 제공되는 유저의 피드백 정보를 활용해 추천시스템을 구축할 수 있다.

결과값
1. 분류 모델링
2. 회귀 모델링
3. 비지도 학습


## 기초 추천시스템 모델링 방법
1. 인기도 기반 추천 시스템  
    조회수, 평균 평점, 리뷰 개수, 좋아요/싫어요 등의 사용자 피드백 정보를 활용해 다른 사용자들에게 가장 인기있는 아이템을 새로운 유저에게 추천하는 방식

2. 연관성 분석  
    - 연관 규칙(Association Rule Analysis)    
    데이터 베이스 내에서 동시에 등장하는 빈도가 높은 컨턴츠의 조합을 찾아내는 방법론

3. 혈업 필터링(CF, Collaborative Filtering)  
    사용자들, 또는 아이템 사이의 유사성을 기반으로 작동하는 추천시스템 방식

4. 기억 기반 (Memory-based) 협업 필터링  
    각각의 사용자들이 각각의 아이템을 선호하는지 여부를 나타낸 사용자-아이템 행렬을 기반으로, 사용자들 사이, 또는 아이템 사이의 유사성을 판단  
    - 사용자-아이템 선호도 행렬: 모든 사용자-아이템 쌍에 대해, 사용자의 아이템에 대한 선호도를 수치로 나타낸 행렬  
    사용자 기반 협업필터링 : 유저들 끼리의 선호도 패턴을 구해 선호도가 비슷한 사람이 긍정평가를 남긴 아이템을 추천한다.  
    아이템 기반 협업필터링 : A 아이템을 좋아하는 유저는 대부분 B를 좋아할때, B 를추천

5. 잠재요인 분석  
    유저들의 아이템에 대한 선호도, 사용자-아이템 행렬을 직접 사용하는 것이 아니라 특정 피쳐에 대한 선호를 나타내는 잠재요인을 찾아내도록 학습하는 방법  
    - 잠재요인: 비지도학습에서 배운 임베딩과 동일한 개념

## 시계열 데이터 (Time Series data)
시계열 데이터 : 시간 순서로 관측된 일련의 데이터.  
순차적 데이터 : 순서를 가지는 데이터(순서가 바뀌면 의미를 잃는다.)  
따라서 시계열 데이터는 순차적 데이터에 포함된다.

시계열 데이터는 규칙 성분과 불규칙 성분으로 나누어진다다.

###규칙 성분(Systematic or Regular Component)  
1. 추세 : 단기적으로는 증감하며 변화하더라도 장기적으로 점차 증가하거나 점차 감소하는 등의 방향성을 보이는 것  
2. 계절성 : 계절에 따른 변화처럼 일정하게, 정기적으로 데이터가 변화 하는 것  
- 성수기 비수기에 따른 비행기 탑승객 수
- 평일/주말에 따른 식당의 손님 수
3. 순환성 : 데이터가 주기적으로 변화하기는 하지만 일정하지 않은 주기로 변하는것  

기술적 분석 : 과거의 데이터를 이해하기 위한 분석과정, 이상치 탐지 및 일반적 행동 설명  
예측적 분석 : 미래의 데이터를 예측하기 위해 과거의 데이터를 분석하는 것  

독립 항등 분호: 어떤 확률변수들이 같은 확률분포로부터 생성된 독립적인 샘플  
머신러닝에서 다루는 대부분의 모델이 IID를 가정하므로 IID를 가정하지 못하는 데이터의 경우 추가적인 고려가 필요.  

자기상관: 한 시계열 내에서 자기자신과 다른 시점과의 상관관계

마르코프 속성 : 랜덤 프로세스의 미래 상태에 대한 조건부 확률 분포가 현재 상태에만 의존하는 것

시계열의 정상성 : 데이터를 생성하는 분포가 시간에 따라 변하지 않고 동일하게 유지  

적용분야에 따른 두가지 접근법
1. 시계열데이터의 성분분해  
시계열 데이터를 각 성분으로 분해하여 예측이나 계절성에 따른 변화를 조정하는데 활용한다.
2. 시간에 따른 변화를 분석
- 이동평균 모델 : p 기간 안의 시계열 데이터를 평균하여 시간 t의 추세-주기를 측정
- 자기회귀 모델 : 시계열의 특정 스텝(t)을 예측하기 위해, 일정 스텝 수(p) 만큼의 과거 데이터들을 입력값으로 선형회귀모델을 학습하는 방식

# 개인 학습
dict_b = {v:k for k, v in dict_a()}  
위의 함수는 dict_a 내부의 key와 value 값을 바꿔주는 역할을 한다.  
해당 함수는 상당히 많이 사용 될 것 같으니까 따로 외워 두자.
